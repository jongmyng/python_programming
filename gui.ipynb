```python
# ttk.Label() 위젯 단순 텍스트 출력 

def label():
    text_label = ttk.Label(text='라벨은 텍스트')
    
    text_label2 = ttk.Label()
    text_label2.configure(text='컨피겨', foreground='white', background='red') #configure로 옵션 수정 
    
    text_label.pack()
    text_label2.pack()
    
    
win = tk.Tk()
win.title('아')
win.geometry('500x200')

label()
win.mainloop()

```

```python
# ttk.Button() 위젯 버튼 만들기
import tkinter as tk
from tkinter import ttk


def input_btn_handler():
    text_label.configure(text='반가워요')  # text_label.configure(수정하는거임)

def Button():
    global text_label    # 전역변수로 선언해야 핸들러 함수에서도 text_label이 수정 가능함 
    text_label = ttk.Label(text='안녕')
    push_btn = ttk.Button(win, text='입력', command = input_btn_handler)
    quit_btn = ttk.Button(win, text='종료')
    
    text_label.pack()
    push_btn.pack()
    quit_btn.pack()



win = tk.Tk()   
win.title('버튼')
win.geometry('500x200')


Button()

win.mainloop()
```

```python
# 객체 이름 = tk.StringVar()             cf) IntVar(): 정수 / DoubleVar() : 실수 
# ttk.Entry(win, textvariable=객체 이름) 
# 한줄로 된 문자열을 입력받음

def input_btn_handler():
    label.configure(text='안녕하세요, ' + name.get()) # 네임을 여기서 받으니까 글로벌해야댐 
    name.set('') # 초기화하기 

    
    
def buildGUI():
    global label # 라벨도 글로벌 넣어야되네.
    
    # 안녕 라벨
    label = ttk.Label(win,text='안녕하세요')
    
    # 텍스트 엔트리 
    global name
    name = tk.StringVar() # 여기로 받고 
    text_entry = ttk.Entry(win, textvariable = name) # 엔트리로 저장함 여기서 
    
    # 입력버튼
    input_btn = ttk.Button(win, text='입력', command = input_btn_handler)
    
    
    #종료 버튼
    quit_btn = ttk.Button(win, text='종료',
                         command = win.destroy) # destroy로 종료하기 
    
    # 팩하기
    label.pack()
    text_entry.pack()
    input_btn.pack()
    quit_btn.pack()
    


win = tk.Tk()
win.title('문자열 + 버튼위젯')
win.geometry('500x200')

buildGUI()
win.mainloop()
```

```python
# Text 위젯  > 여러 줄로 된 텍스트를 입력 받음 
# 객체이름 = tk.Text(부모윈도우, width = 폭 , height = 높이, wrap = 줄바꿈처리방식)

import tkinter as tk
from tkinter import ttk

def input_btn_handler():
    print(text.get(0.0, tk.END)) #0.0 첫문자열 다 가져오기
    text.delete(0.0, tk.END)   # 0.0부터 다 삭제하기 


def buildGUI():
    global text
    
    # 텍스트 위젯달기 
    text = tk.Text(win, width = 30, height = 5, wrap =tk.WORD) # 단어 단위로 줄바꿈 
    
    # 출력버튼 
    input_btn = ttk.Button(win, text='출력', command= input_btn_handler)
    
    # 팩하기
    text.pack()
    input_btn.pack()

win = tk.Tk()
win.title('텍스트 + 버튼위젯')
buildGUI()
win.mainloop()
```

```python
# scrolledtext 위젯
# 얘는 따로 임포트해서써야되네 
import tkinter as tk
from tkinter import ttk
from tkinter import scrolledtext


def buildGUI():
    global text
    
    # 텍스트 위젯달기 
    text = scrolledtext.ScrolledText(win, width=30, height=5, wrap=tk.WORD)
    
    text.pack()
    
win = tk.Tk()
win.title('텍스트 스크롤')
buildGUI()
win.mainloop()
```

```python
# 체크버튼 위젯 
# tk변수객체이름 = tk.IntVar()
# 객체이름 = ttk.Checkbutton(win, text='aadfa', variable=tk변수객체이름, command=핸들러함수)

import tkinter as tk
from tkinter import ttk
from tkinter import messagebox


def buildGUI():
    global check
    check = tk.IntVar()
    checkbox = ttk.Checkbutton(win, text='옵션선택하기', variable=check, command=open_dialog_box)
    
    #팩하기
    checkbox.pack()

def open_dialog_box():
    if check.get() ==1:
        messagebox.showinfo('확인', '옵션선택')
    else:
        messagebox.showinfo('확인', '옵션해제')
win = tk.Tk()
win.title('체크버튼')
win.geometry('500x200')

buildGUI()
win.mainloop()
```

```python
# Radiobutton 위젯 - 여러 항목중에 하나만 택하기 
# tk변수이름 = tk.IntVar()
# 객체이름 = ttk.RadioButton(win, text='문자열', variable = tk변수객체이름, value=각버튼의식별아이디값, command=)

import tkinter as tk
from tkinter import ttk

GENDERS = ['남성', '여성', '기타']
COLORS =['red', 'yellow', 'purple']

  
    
def radio_btn_handler():
    choice = gender.get()
    win.configure(background=COLORS[choice])

def buildGUI():
    # 성별은?
    text_label = ttk.Label(text='당신의 성별은?')
    text_label.pack()
    
    # 젠더선택 
    global gender 
    gender = tk.IntVar() # 숫자 받음 
    for i in range(3):
        radio_btn = ttk.Radiobutton(win, text=GENDERS[i],
                                   value=i,
                                   variable=gender, 
                                   command=radio_btn_handler) # 여기 잘 모르겠는데 일단 넘어가 
        
        radio_btn.pack() # 팩을 반복문 안에 넣어야 3번하지
    
    gender.set(-1) # 이건 잘모르겠네

win = tk.Tk()
win.title('라디오체크')
win.geometry('500x200')

buildGUI()
win.mainloop()
```

```python
# frame을 활용해서 버튼의 위치를 정해주기 

def BuildGUI():
    btn1 = ttk.Button(win, text='버튼1')
    btn2 = ttk.Button(win, text= '매우긴 ~ 너미아러미ㅏㅓㅇㄹ마러ㅣ\n')
    
    btn_group = ttk.Frame(win)
    btn3 = ttk.Button(btn_group, text='버튼3')
    btn4 = ttk.Button(btn_group, text='버튼4')
    
    btn1.grid(row=0, column = 0)
    btn2.grid(row=1,column=0)
    btn3.pack()
    btn4.pack()
    btn_group.grid(row=1, column = 1 , sticky='w')
    
win = tk.Tk()
win.title('frame')
win.geometry('500x200')

BuildGUI()
win.mainloop()
```
